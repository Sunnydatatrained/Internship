```python
import pandas as pd
import numpy as np
```

```python
zomato_data = pd.read_csv('https://github.com/dsrscientist/dataset4/raw/main/zomato.csv', encoding='latin1')
```

```python
zomato_data
```

```python
# Load the country code dataset
country_code_data = pd.read_excel('https://github.com/dsrscientist/dataset4/raw/main/Country-Code.xlsx')
```

```python
country_code_data
```

```python
zomato_data = zomato_data.merge(country_code_data, on='Country Code', how='left')
```

```python
zomato_data
```

```python
zomato_data.info()
```

```python
# Handle missing values
zomato_data['Cuisines'].fillna('', inplace=True)  # Fill missing Cuisines with an empty string or any suitable default value
```

```python
zomato_data.info()
```

```python
# Convert binary columns to 1s and 0s
binary_columns = ['Has Table booking', 'Has Online delivery', 'Is delivering now', 'Switch to order menu']
for column in binary_columns:
    zomato_data[column] = (zomato_data[column] == 'Yes').astype(int)
```

```python
zomato_data.info()
```

```python
# Preprocess the "Cuisines" column
def preprocess_cuisines(cuisines):
    cuisines = cuisines.split(', ')
    cuisines = [cuisine.strip() for cuisine in cuisines]
    return cuisines

zomato_data['Cuisines'] = zomato_data['Cuisines'].apply(preprocess_cuisines)
```

```python
# Feature selection
selected_features = [
    'Has Table booking', 'Has Online delivery', 'Is delivering now', 'Switch to order menu',
    'Price range', 'Aggregate rating', 'Votes'
]
X = zomato_data[selected_features]
y_avg_cost = zomato_data['Average Cost for two']
y_price_range = zomato_data['Price range']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression  # Example for regression
from sklearn.ensemble import RandomForestClassifier  # Example for classification
```

```python
# Split the data into training and testing sets
X_train, X_test, y_avg_cost_train, y_avg_cost_test = train_test_split(X, y_avg_cost, test_size=0.2, random_state=42)
X_train, X_test, y_price_range_train, y_price_range_test = train_test_split(X, y_price_range, test_size=0.2, random_state=42)
```

```python
# Create and train regression model
regression_model = LinearRegression()
regression_model.fit(X_train, y_avg_cost_train)
```

```python
# Create and train classification model
classification_model = RandomForestClassifier()
classification_model.fit(X_train, y_price_range_train)
```

```python
from sklearn.metrics import mean_absolute_error, accuracy_score, f1_score
```

```python
# Evaluate regression model
y_avg_cost_pred = regression_model.predict(X_test)
mae = mean_absolute_error(y_avg_cost_test, y_avg_cost_pred)
```

```python
# Evaluate classification model
y_price_range_pred = classification_model.predict(X_test)
accuracy = accuracy_score(y_price_range_test, y_price_range_pred)
f1 = f1_score(y_price_range_test, y_price_range_pred, average='weighted')
```

```python
# Make predictions
y_avg_cost_pred = regression_model.predict(X_test)
y_price_range_pred = classification_model.predict(X_test)
```

```python
# Example conclusion
print("Regression Model - Mean Absolute Error:", mae)
print("Classification Model - Accuracy:", accuracy)
print("Classification Model - F1-score:", f1)
```
